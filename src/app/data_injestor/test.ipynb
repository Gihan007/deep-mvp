```python
import requests
import pandas as pd
import time
from tabulate import tabulate

# Alpha Vantage API Key
API_KEY = 'NV4M8POQQLMF2B4Z'  # Replace with your API key

def fetch_company_overview(ticker):     # get - SharesOutstanding, MarketCapitalization
    """Fetch company overview data"""
    url = f'https://www.alphavantage.co/query?function=OVERVIEW&symbol={ticker}&apikey={API_KEY}'
    response = requests.get(url)
    return response.json()

def fetch_income_statement(ticker):     # get - Net Income.  | nnualReports -> (operatingIncome)
    """Fetch income statement data"""
    url = f'https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol={ticker}&apikey={API_KEY}'
    response = requests.get(url)
    return response.json()

def fetch_balance_sheet(ticker):        # get -  annualReports -> (totalAssets, totalCurrentLiabilities)
    """Fetch balance sheet data"""
    url = f'https://www.alphavantage.co/query?function=BALANCE_SHEET&symbol={ticker}&apikey={API_KEY}'
    response = requests.get(url)
    return response.json()

def fetch_current_price(ticker):      # stock price
    """Fetch current stock price"""
    url = f'https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol={ticker}&apikey={API_KEY}'
    response = requests.get(url)
    data = response.json()
    if 'Global Quote' in data and '05. price' in data['Global Quote']:
        return float(data['Global Quote']['05. price'])
    return None

#.  ---------------------------------------------------------------- #
#.  ---------------------------------------------------------------- #
def calculate_roic_5y(income_data, balance_data):
    """Calculate 5-year average ROIC"""
    try:
        annual_income = income_data.get('annualReports', [])[:5]
        annual_balance = balance_data.get('annualReports', [])[:5]
        
        roic_values = []
        for income, balance in zip(annual_income, annual_balance):
            nopat = float(income.get('operatingIncome', 0))
            total_assets = float(balance.get('totalAssets', 0))
            current_liabilities = float(balance.get('totalCurrentLiabilities', 0))
            invested_capital = total_assets - current_liabilities
            
            if invested_capital > 0:
                roic = (nopat / invested_capital) * 100
                roic_values.append(roic)
        
        print("calculate_roic_5y OK")
        return sum(roic_values) / len(roic_values) if roic_values else 0
    except:
        return 0

def calculate_net_income_5y_avg(income_data):
    """Calculate 5-year average net income"""
    try:
        annual_reports = income_data.get('annualReports', [])[:5]
        net_incomes = [float(report.get('netIncome', 0)) for report in annual_reports]
        return sum(net_incomes) / len(net_incomes) if net_incomes else 0
    except:
        return 0

def calculate_intrinsic_value(net_income_5y_avg, growth_rate=0.05, discount_rate=0.10, terminal_multiple=15):
    """Calculate intrinsic value using simplified DCF"""
    try:
        # Project 5 years of cash flows
        cash_flows = []
        for year in range(1, 6):
            cf = net_income_5y_avg * ((1 + growth_rate) ** year)
            pv = cf / ((1 + discount_rate) ** year)
            cash_flows.append(pv)
        
        # Terminal value
        terminal_cf = net_income_5y_avg * ((1 + growth_rate) ** 5)
        terminal_value = terminal_cf * terminal_multiple
        pv_terminal = terminal_value / ((1 + discount_rate) ** 5)
        
        return sum(cash_flows) + pv_terminal
    except:
        return 0

def fetch_company_data(ticker):
    """Fetch all required data for a company"""
    print(f"Fetching data for {ticker}...")
    
    try:
        # Fetch data
        overview = fetch_company_overview(ticker)
        #time.sleep(12)  # Alpha Vantage free tier: 5 calls per minute
        income = fetch_income_statement(ticker)
        #time.sleep(12)
        balance = fetch_balance_sheet(ticker)
        #time.sleep(12)
        stock_price = fetch_current_price(ticker)
        #time.sleep(12)
        

        
        # Extract data
        shares_outstanding = float(overview.get('SharesOutstanding', 0))
        market_cap = float(overview.get('MarketCapitalization', 0))
        
        # Calculate metrics
        roic_5y = calculate_roic_5y(income, balance)
        net_income_5y_avg = calculate_net_income_5y_avg(income)
        intrinsic_value = calculate_intrinsic_value(net_income_5y_avg)
        
        # Calculate Earnings Yield
        if shares_outstanding > 0 and stock_price:
            earnings_per_share = net_income_5y_avg / shares_outstanding
            earnings_yield = (earnings_per_share / stock_price) * 100
        else:
            earnings_yield = 0
        
        # Calculate Intrinsic to Market Cap ratio
        intrinsic_to_market_cap = (intrinsic_value / market_cap * 100) if market_cap > 0 else 0
        
        return {
            'Ticker': ticker,
            'ROIC_5Y': roic_5y,
            'Earnings_Yield': earnings_yield,
            'Intrinsic_to_Market_Cap': intrinsic_to_market_cap,
            'Stock_Price': stock_price,
            'Market_Cap': market_cap
        }
    
    except Exception as e:
        print(f"Error fetching data for {ticker}: {str(e)}")
        return None

def calculate_rankings(data_list):
    """Calculate rankings for each metric"""
    df = pd.DataFrame(data_list)
    
    # Rank each metric (highest value = rank 1)
    df['ROIC_Rank'] = df['ROIC_5Y'].rank(ascending=False, method='min').astype(int)
    df['EY_Rank'] = df['Earnings_Yield'].rank(ascending=False, method='min').astype(int)
    df['IM_Rank'] = df['Intrinsic_to_Market_Cap'].rank(ascending=False, method='min').astype(int)
    
    # Calculate overall rank (lowest sum = rank 1)
    df['Rank_Sum'] = df['ROIC_Rank'] + df['EY_Rank'] + df['IM_Rank']
    df['Overall_Rank'] = df['Rank_Sum'].rank(method='min').astype(int)
    
    # Sort by overall rank
    df = df.sort_values('Overall_Rank')
    
    return df

def format_table(df):
    """Format the dataframe for display"""
    display_df = pd.DataFrame()
    
    display_df['Overall Rank'] = df['Overall_Rank']
    display_df['Ticker'] = df['Ticker']
    display_df['ROIC 5Y Avg'] = df.apply(
        lambda row: f"{row['ROIC_5Y']:.2f}% ({row['ROIC_Rank']})", axis=1
    )
    display_df['Earnings Yield'] = df.apply(
        lambda row: f"{row['Earnings_Yield']:.2f}% ({row['EY_Rank']})", axis=1
    )
    display_df['Intrinsic/Market Cap'] = df.apply(
        lambda row: f"{row['Intrinsic_to_Market_Cap']:.2f}% ({row['IM_Rank']})", axis=1
    )
    
    return display_df

def main():
    # Input tickers

    tickers = ['AAPL','MSFT','GOOGL']
    
    print(f"\nFetching data for {len(tickers)} companies...")
    print("Note: This may take several minutes due to API rate limits\n")
    
    # Fetch data for all companies
    company_data = []
    for ticker in tickers:
        data = fetch_company_data(ticker)
        if data:
            company_data.append(data)
    
    if not company_data:
        print("No data was successfully fetched. Please check your API key and tickers.")
        return
    
    # Calculate rankings
    ranked_df = calculate_rankings(company_data)
    
    # Format and display table
    display_df = format_table(ranked_df)
    
    print("\n" + "="*100)
    print("FINANCIAL RANKING TABLE")
    print("="*100 + "\n")
    print(tabulate(display_df, headers='keys', tablefmt='grid', showindex=False))
    print("\n" + "="*100)
    print("Legend: Value (Rank) - Lower rank number = Better performance")
    print("Overall Rank: Sum of all three ranks (lower is better)")
    print("="*100 + "\n")

if __name__ == "__main__":
    main()
```

```python
from datetime import datetime
from neo4j import GraphDatabase
import math
import requests

NEO4J_URI = "bolt://localhost:7687"
NEO4J_USERNAME = "neo4j"
NEO4J_PASSWORD = "your_neo4j_password_here"
ALPHA_VANTAGE_API_KEY = 'K762T7R2D05LK8Y7'

driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USERNAME, NEO4J_PASSWORD))

def get_last_5yr_avg_ROIC(driver, ticker):
    metric_name = "NetOperatingProfitAfterTaxes"
    current_year = datetime.now().year
    years = [current_year - i for i in range(1, 6)]
    year_props = [f"year_{y}" for y in years]
    return_fields = ", ".join([f"m.{p} AS {p}" for p in year_props])
    query = f"""
        MATCH (m:Metric {{ ticker: $ticker, metricName: $metric_name }})
        RETURN {return_fields}
    """
    with driver.session() as session:
        row = session.run(query, ticker=ticker, metric_name=metric_name).single()
    if not row:
        return None
    values = []
    for p in year_props:
        val = row.get(p)
        if val is not None and not (isinstance(val, float) and math.isnan(val)):
            values.append(float(val))

    if not values:
        return None
    return sum(values) / len(values)


def get_last_5yr_avg_NetIncome(driver, ticker):
    metric_name = "NetIncome"
    current_year = datetime.now().year
    years = [current_year - i for i in range(1, 6)]
    year_props = [f"year_{y}" for y in years]
    return_fields = ", ".join([f"m.{p} AS {p}" for p in year_props])
    query = f"""
        MATCH (m:Metric {{ ticker: $ticker, metricName: $metric_name }})
        RETURN {return_fields}
    """
    with driver.session() as session:
        row = session.run(query, ticker=ticker, metric_name=metric_name).single()
    if not row:
        return None
    values = []
    for p in year_props:
        val = row.get(p)
        if val is not None and not (isinstance(val, float) and math.isnan(val)):
            values.append(float(val))
    if not values:
        return None
    return sum(values) / len(values)


def fetch_company_overview(ticker):
    """Fetch shares outstanding and market cap from Alpha Vantage"""
    url = f'https://www.alphavantage.co/query?function=OVERVIEW&symbol={ticker}&apikey={ALPHA_VANTAGE_API_KEY}'
    response = requests.get(url)
    data = response.json()
    
    shares_outstanding = float(data.get('SharesOutstanding', 0)) if data.get('SharesOutstanding') else None
    market_cap = float(data.get('MarketCapitalization', 0)) if data.get('MarketCapitalization') else None
    
    return {
        "shares_outstanding": shares_outstanding,
        "market_cap": market_cap
    }


def fetch_current_stock_price(ticker):
    """Fetch current stock price"""
    url = f'https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol={ticker}&apikey={ALPHA_VANTAGE_API_KEY}'
    response = requests.get(url)
    data = response.json()
    
    if 'Global Quote' in data and '05. price' in data['Global Quote']:
        return float(data['Global Quote']['05. price'])
    return None


def get_Intrinsic_value(driver, ticker):
    """
    Get intrinsic value (Equity Value) from ValuationSummary metric in Neo4j
    """
    metric_name = "ValuationSummary"
    query = """
        MATCH (m:Metric {ticker: $ticker, metricName: $metric_name})
        RETURN m.EquityValue AS equity_value
    """
    with driver.session() as session:
        row = session.run(query, ticker=ticker, metric_name=metric_name).single()
    if not row:
        return None
    equity_value = row.get('equity_value')
    # Check if value exists and is valid (not None or NaN)
    if equity_value is not None and not (isinstance(equity_value, float) and math.isnan(equity_value)):
        return float(equity_value)
    return None


def calculate_metrics_for_tickers(tickers):
    """
    Calculate metrics for all tickers and return ranked results
    """
    results = []
    
    # Step 1: Gather data for all tickers
    for ticker in tickers:
        try:
            print(f"Processing {ticker}...")
            
            # Get ROIC 5Y Avg
            roic_5y_avg = get_last_5yr_avg_ROIC(driver, ticker)
            
            # Get Net Income 5Y Avg
            net_income_5y_avg = get_last_5yr_avg_NetIncome(driver, ticker)
            
            # Get company overview (shares outstanding, market cap)
            company_data = fetch_company_overview(ticker)
            shares_outstanding = company_data['shares_outstanding']
            market_cap = company_data['market_cap']
            
            # Get current stock price
            current_price = fetch_current_stock_price(ticker)
            
            # Get intrinsic value
            intrinsic_value = get_Intrinsic_value(ticker)
            
            # Calculate Earnings Yield
            earnings_yield = None
            if net_income_5y_avg and shares_outstanding and current_price and current_price > 0:
                eps_5y_avg = net_income_5y_avg / shares_outstanding
                earnings_yield = eps_5y_avg / current_price
            
            # Calculate Intrinsic Value to Market Cap ratio
            intrinsic_to_mc = None
            if intrinsic_value and market_cap and market_cap > 0:
                intrinsic_to_mc = intrinsic_value / market_cap
            
            results.append({
                'ticker': ticker,
                'roic_5y_avg': roic_5y_avg,
                'earnings_yield': earnings_yield,
                'intrinsic_to_mc': intrinsic_to_mc
            })
            
        except Exception as e:
            print(f"Error processing {ticker}: {e}")
            results.append({
                'ticker': ticker,
                'roic_5y_avg': None,
                'earnings_yield': None,
                'intrinsic_to_mc': None
            })
    
    # Step 2: Rank each metric (higher is better, so rank 1 = highest value)
    # Filter out None values for ranking
    
    # Rank ROIC
    valid_roic = [(i, r['roic_5y_avg']) for i, r in enumerate(results) if r['roic_5y_avg'] is not None]
    valid_roic.sort(key=lambda x: x[1], reverse=True)  # Sort descending
    roic_ranks = {}
    for rank, (idx, _) in enumerate(valid_roic, 1):
        roic_ranks[idx] = rank
    
    # Rank Earnings Yield
    valid_ey = [(i, r['earnings_yield']) for i, r in enumerate(results) if r['earnings_yield'] is not None]
    valid_ey.sort(key=lambda x: x[1], reverse=True)
    ey_ranks = {}
    for rank, (idx, _) in enumerate(valid_ey, 1):
        ey_ranks[idx] = rank
    
    # Rank Intrinsic to MC
    valid_imc = [(i, r['intrinsic_to_mc']) for i, r in enumerate(results) if r['intrinsic_to_mc'] is not None]
    valid_imc.sort(key=lambda x: x[1], reverse=True)
    imc_ranks = {}
    for rank, (idx, _) in enumerate(valid_imc, 1):
        imc_ranks[idx] = rank
    
    # Step 3: Calculate overall rank and format output
    final_results = []
    for i, result in enumerate(results):
        roic_rank = roic_ranks.get(i)
        ey_rank = ey_ranks.get(i)
        imc_rank = imc_ranks.get(i)
        
        # Calculate overall rank score (lower is better)
        overall_score = None
        rank_components = [r for r in [roic_rank, ey_rank, imc_rank] if r is not None]
        if rank_components:
            overall_score = sum(rank_components)
        
        final_results.append({
            'ticker': result['ticker'],
            'roic_5y_avg': result['roic_5y_avg'],
            'roic_rank': roic_rank,
            'earnings_yield': result['earnings_yield'],
            'earnings_yield_rank': ey_rank,
            'intrinsic_to_mc': result['intrinsic_to_mc'],
            'intrinsic_to_mc_rank': imc_rank,
            'overall_score': overall_score
        })
    
    # Sort by overall score (lower is better)
    final_results.sort(key=lambda x: x['overall_score'] if x['overall_score'] is not None else float('inf'))
    
    # Assign overall ranks
    for overall_rank, item in enumerate(final_results, 1):
        item['overall_rank'] = overall_rank if item['overall_score'] is not None else None
    
    return final_results


def format_output(results):
    """Format results for display"""
    output = []
    for r in results:
        output.append({
            'Overall Rank': r['overall_rank'],
            'Ticker': r['ticker'],
            'ROIC 5Y Avg': f"{r['roic_5y_avg']:.4f} ({r['roic_rank']})" if r['roic_5y_avg'] is not None else "N/A",
            'Earnings Yield': f"{r['earnings_yield']:.4f} ({r['earnings_yield_rank']})" if r['earnings_yield'] is not None else "N/A",
            'Intrinsic to Market Cap': f"{r['intrinsic_to_mc']:.4f} ({r['intrinsic_to_mc_rank']})" if r['intrinsic_to_mc'] is not None else "N/A"
        })
    return output


# Main execution
if __name__ == "__main__":
    tickers = ['AAPL', 'MSFT', 'GOOGL']
    
    results = calculate_metrics_for_tickers(tickers)
    formatted_results = format_output(results)
    
    # Print results
    print("\n" + "="*80)
    print("STOCK RANKING RESULTS")
    print("="*80)
    for item in formatted_results:
        print(f"\nOverall Rank: {item['Overall Rank']}")
        print(f"Ticker: {item['Ticker']}")
        print(f"ROIC 5Y Avg: {item['ROIC 5Y Avg']}")
        print(f"Earnings Yield: {item['Earnings Yield']}")
        print(f"Intrinsic to Market Cap: {item['Intrinsic to Market Cap']}")
        print("-"*80)
    
    # Close driver
    driver.close()
```

```python
from datetime import datetime, timedelta
from neo4j import GraphDatabase
import math
import requests
import time

NEO4J_URI = "bolt://localhost:7687"
NEO4J_USERNAME = "neo4j"
NEO4J_PASSWORD = "your_neo4j_password_here"
ALPHA_VANTAGE_API_KEY = 'K762T7R2D05LK8Y7'

driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USERNAME, NEO4J_PASSWORD))

# ============================================================================
# CACHE MANAGEMENT FUNCTIONS
# ============================================================================

def create_cache_node_constraint(driver):
    """Create unique constraint for cache nodes (run once during setup)"""
    with driver.session() as session:
        try:
            session.run("""
                CREATE CONSTRAINT stock_cache_ticker IF NOT EXISTS
                FOR (c:StockCache) REQUIRE c.ticker IS UNIQUE
            """)
            print("Cache constraint created successfully")
        except Exception as e:
            print(f"Constraint may already exist: {e}")


def store_stock_cache(driver, ticker, price=None, shares_outstanding=None, market_cap=None):
    """Store or update stock cache data in Neo4j"""
    timestamp = datetime.now().isoformat()
    cache_date = datetime.now().date().isoformat()  # Store just the date
    
    query = """
    MERGE (c:StockCache {ticker: $ticker})
    SET c.timestamp = $timestamp,
        c.cache_date = $cache_date
    """
    params = {
        'ticker': ticker,
        'timestamp': timestamp,
        'cache_date': cache_date
    }
    
    if price is not None:
        query += ", c.price = $price"
        params['price'] = price
    
    if shares_outstanding is not None:
        query += ", c.shares_outstanding = $shares_outstanding"
        params['shares_outstanding'] = shares_outstanding
    
    if market_cap is not None:
        query += ", c.market_cap = $market_cap"
        params['market_cap'] = market_cap
    
    with driver.session() as session:
        session.run(query, params)
    
    print(f"Cached data for {ticker} at {timestamp}")


def get_cached_stock_data(driver, ticker):
    """
    Retrieve cached stock data if it exists and is from today
    Returns dict with data or None if cache miss/outdated
    If outdated, automatically deletes the cache node
    """
    query = """
    MATCH (c:StockCache {ticker: $ticker})
    RETURN c.price AS price,
           c.shares_outstanding AS shares_outstanding,
           c.market_cap AS market_cap,
           c.timestamp AS timestamp,
           c.cache_date AS cache_date
    """
    
    with driver.session() as session:
        result = session.run(query, ticker=ticker).single()
    
    if not result:
        print(f"Cache MISS for {ticker} - no data found")
        return None
    
    # Check if cache is from today
    today = datetime.now().date().isoformat()
    cache_date = result['cache_date']
    
    if cache_date != today:
        print(f"Cache OUTDATED for {ticker} - cached on: {cache_date}, today: {today}")
        # Delete the outdated cache node
        delete_query = """
        MATCH (c:StockCache {ticker: $ticker})
        DELETE c
        """
        with driver.session() as session:
            session.run(delete_query, ticker=ticker)
        print(f"Deleted outdated cache for {ticker}")
        return None
    
    print(f"Cache HIT for {ticker} - cached today at {result['timestamp']}")
    return {
        'price': result['price'],
        'shares_outstanding': result['shares_outstanding'],
        'market_cap': result['market_cap'],
        'timestamp': result['timestamp']
    }



def clear_outdated_cache(driver):
    """Clear cache entries that are not from today"""
    today = datetime.now().date().isoformat()
    
    query = """
    MATCH (c:StockCache)
    WHERE c.cache_date < $today
    DELETE c
    RETURN count(c) AS deleted_count
    """
    
    with driver.session() as session:
        result = session.run(query, today=today).single()
        deleted = result['deleted_count']
    
    print(f"Cleared {deleted} outdated cache entries (not from today)")
    return deleted


def clear_all_cache(driver):
    """Clear all cache entries (useful for midnight cleanup)"""
    query = """
    MATCH (c:StockCache)
    DELETE c
    RETURN count(c) AS deleted_count
    """
    
    with driver.session() as session:
        result = session.run(query).single()
        deleted = result['deleted_count']
    
    print(f"Cleared ALL cache: {deleted} entries deleted")
    return deleted


# ============================================================================
# ALPHA VANTAGE API FUNCTIONS (with caching)
# ============================================================================

def fetch_company_overview_cached(driver, ticker):
    """
    Fetch shares outstanding and market cap - uses cache first
    """
    # Try cache first
    cached = get_cached_stock_data(driver, ticker)
    if cached and cached['shares_outstanding'] and cached['market_cap']:
        return {
            'shares_outstanding': cached['shares_outstanding'],
            'market_cap': cached['market_cap']
        }
    
    # Cache miss - fetch from API
    print(f"Fetching company overview from API for {ticker}...")
    url = f'https://www.alphavantage.co/query?function=OVERVIEW&symbol={ticker}&apikey={ALPHA_VANTAGE_API_KEY}'
    
    try:
        response = requests.get(url)
        data = response.json()
        
        shares_outstanding = float(data.get('SharesOutstanding', 0)) if data.get('SharesOutstanding') else None
        market_cap = float(data.get('MarketCapitalization', 0)) if data.get('MarketCapitalization') else None
        
        # Store in cache
        store_stock_cache(driver, ticker, 
                         shares_outstanding=shares_outstanding, 
                         market_cap=market_cap)
        
        # Rate limit protection
        time.sleep(2)  # Alpha Vantage free tier: 5 calls/min
        
        return {
            'shares_outstanding': shares_outstanding,
            'market_cap': market_cap
        }
    except Exception as e:
        print(f"Error fetching company overview for {ticker}: {e}")
        return {'shares_outstanding': None, 'market_cap': None}


def fetch_current_stock_price_cached(driver, ticker):
    """
    Fetch current stock price - uses cache first
    """
    # Try cache first
    cached = get_cached_stock_data(driver, ticker)
    if cached and cached['price']:
        return cached['price']
    
    # Cache miss - fetch from API
    print(f"Fetching stock price from API for {ticker}...")
    url = f'https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol={ticker}&apikey={ALPHA_VANTAGE_API_KEY}'
    
    try:
        response = requests.get(url)
        data = response.json()
        
        price = None
        if 'Global Quote' in data and '05. price' in data['Global Quote']:
            price = float(data['Global Quote']['05. price'])
        
        # Store in cache
        if price:
            store_stock_cache(driver, ticker, price=price)
        
        # Rate limit protection
        time.sleep(2)
        
        return price
    except Exception as e:
        print(f"Error fetching price for {ticker}: {e}")
        return None


# ============================================================================
# EXISTING NEO4J FUNCTIONS (unchanged)
# ============================================================================

def get_last_5yr_avg_ROIC(driver, ticker):
    metric_name = "NetOperatingProfitAfterTaxes"
    current_year = datetime.now().year
    years = [current_year - i for i in range(1, 6)]
    year_props = [f"year_{y}" for y in years]
    return_fields = ", ".join([f"m.{p} AS {p}" for p in year_props])
    query = f"""
        MATCH (m:Metric {{ ticker: $ticker, metricName: $metric_name }})
        RETURN {return_fields}
    """
    with driver.session() as session:
        row = session.run(query, ticker=ticker, metric_name=metric_name).single()
    if not row:
        return None
    values = []
    for p in year_props:
        val = row.get(p)
        if val is not None and not (isinstance(val, float) and math.isnan(val)):
            values.append(float(val))

    if not values:
        return None
    return sum(values) / len(values)


def get_last_5yr_avg_NetIncome(driver, ticker):
    metric_name = "NetIncome"
    current_year = datetime.now().year
    years = [current_year - i for i in range(1, 6)]
    year_props = [f"year_{y}" for y in years]
    return_fields = ", ".join([f"m.{p} AS {p}" for p in year_props])
    query = f"""
        MATCH (m:Metric {{ ticker: $ticker, metricName: $metric_name }})
        RETURN {return_fields}
    """
    with driver.session() as session:
        row = session.run(query, ticker=ticker, metric_name=metric_name).single()
    if not row:
        return None
    values = []
    for p in year_props:
        val = row.get(p)
        if val is not None and not (isinstance(val, float) and math.isnan(val)):
            values.append(float(val))
    if not values:
        return None
    return sum(values) / len(values)



def get_Intrinsic_value(driver, ticker):
    """
    Get intrinsic value (Equity Value) from ValuationSummary metric in Neo4j
    """
    metric_name = "ValuationSummary"
    query = """
        MATCH (m:Metric {ticker: $ticker, metricName: $metric_name})
        RETURN m.EquityValue AS equity_value
    """
    with driver.session() as session:
        row = session.run(query, ticker=ticker, metric_name=metric_name).single()
    if not row:
        return None
    equity_value = row.get('equity_value')
    # Check if value exists and is valid (not None or NaN)
    if equity_value is not None and not (isinstance(equity_value, float) and math.isnan(equity_value)):
        return float(equity_value)
    return None



# ============================================================================
# METRICS CALCULATION (updated to use cache)
# ============================================================================

def calculate_metrics_for_tickers(tickers):
    """
    Calculate metrics for all tickers and return ranked results
    Uses caching to minimize API calls
    """
    results = []
    
    # Clear outdated cache before starting (optional)
    clear_outdated_cache(driver)
    
    # Step 1: Gather data for all tickers
    for ticker in tickers:
        try:
            print(f"\n{'='*60}")
            print(f"Processing {ticker}...")
            print('='*60)
            
            # Get ROIC 5Y Avg from Neo4j
            roic_5y_avg = get_last_5yr_avg_ROIC(driver, ticker)
            
            # Get Net Income 5Y Avg from Neo4j
            net_income_5y_avg = get_last_5yr_avg_NetIncome(driver, ticker)
            
            # Get company overview (with cache)
            company_data = fetch_company_overview_cached(driver, ticker)
            shares_outstanding = company_data['shares_outstanding']
            market_cap = company_data['market_cap']
            
            # Get current stock price (with cache)
            current_price = fetch_current_stock_price_cached(driver, ticker)
            
            # Get intrinsic value
            intrinsic_value = get_Intrinsic_value(driver, ticker)
            
            # Calculate Earnings Yield
            earnings_yield = None
            if net_income_5y_avg and shares_outstanding and current_price and current_price > 0:
                eps_5y_avg = net_income_5y_avg / shares_outstanding
                earnings_yield = eps_5y_avg / current_price
            
            # Calculate Intrinsic Value to Market Cap ratio
            intrinsic_to_mc = None
            if intrinsic_value and market_cap and market_cap > 0:
                intrinsic_to_mc = intrinsic_value / market_cap
            
            results.append({
                'ticker': ticker,
                'roic_5y_avg': roic_5y_avg,
                'earnings_yield': earnings_yield,
                'intrinsic_to_mc': intrinsic_to_mc
            })
            
        except Exception as e:
            print(f"Error processing {ticker}: {e}")
            results.append({
                'ticker': ticker,
                'roic_5y_avg': None,
                'earnings_yield': None,
                'intrinsic_to_mc': None
            })
    

    print("results: ", results)

    # Step 2: Rank each metric
    valid_roic = [(i, r['roic_5y_avg']) for i, r in enumerate(results) if r['roic_5y_avg'] is not None]
    valid_roic.sort(key=lambda x: x[1], reverse=True)
    roic_ranks = {}
    for rank, (idx, _) in enumerate(valid_roic, 1):
        roic_ranks[idx] = rank
    
    valid_ey = [(i, r['earnings_yield']) for i, r in enumerate(results) if r['earnings_yield'] is not None]
    valid_ey.sort(key=lambda x: x[1], reverse=True)
    ey_ranks = {}
    for rank, (idx, _) in enumerate(valid_ey, 1):
        ey_ranks[idx] = rank
    
    valid_imc = [(i, r['intrinsic_to_mc']) for i, r in enumerate(results) if r['intrinsic_to_mc'] is not None]
    valid_imc.sort(key=lambda x: x[1], reverse=True)
    imc_ranks = {}
    for rank, (idx, _) in enumerate(valid_imc, 1):
        imc_ranks[idx] = rank
    
    # Step 3: Calculate overall rank
    final_results = []
    for i, result in enumerate(results):
        roic_rank = roic_ranks.get(i)
        ey_rank = ey_ranks.get(i)
        imc_rank = imc_ranks.get(i)
        
        overall_score = None
        rank_components = [r for r in [roic_rank, ey_rank, imc_rank] if r is not None]
        if rank_components:
            overall_score = sum(rank_components)
        
        final_results.append({
            'ticker': result['ticker'],
            'roic_5y_avg': result['roic_5y_avg'],
            'roic_rank': roic_rank,
            'earnings_yield': result['earnings_yield'],
            'earnings_yield_rank': ey_rank,
            'intrinsic_to_mc': result['intrinsic_to_mc'],
            'intrinsic_to_mc_rank': imc_rank,
            'overall_score': overall_score
        })
    
    final_results.sort(key=lambda x: x['overall_score'] if x['overall_score'] is not None else float('inf'))
    
    for overall_rank, item in enumerate(final_results, 1):
        item['overall_rank'] = overall_rank if item['overall_score'] is not None else None
    
    return final_results


def format_output(results):
    """Format results for display"""
    output = []
    for r in results:
        output.append({
            'Overall Rank': r['overall_rank'],
            'Ticker': r['ticker'],
            'ROIC 5Y Avg': f"{r['roic_5y_avg']:.4f} ({r['roic_rank']})" if r['roic_5y_avg'] is not None else "N/A",
            'Earnings Yield': f"{r['earnings_yield']:.4f} ({r['earnings_yield_rank']})" if r['earnings_yield'] is not None else "N/A",
            'Intrinsic to Market Cap': f"{r['intrinsic_to_mc']:.4f} ({r['intrinsic_to_mc_rank']})" if r['intrinsic_to_mc'] is not None else "N/A"
        })
    return output


# ============================================================================
# SCHEDULED CLEANUP (for midnight cache clearing)
# ============================================================================

def schedule_midnight_cleanup():
    """
    Run this as a scheduled task (cron job) at midnight to clear cache
    Example cron: 0 0 * * * /usr/bin/python3 /path/to/script.py --cleanup
    """
    print(f"Running midnight cleanup at {datetime.now()}")
    deleted = clear_all_cache(driver)
    print(f"Cleanup complete: {deleted} entries removed")


# ============================================================================
# MAIN EXECUTION
# ============================================================================

if __name__ == "__main__":
    import sys
    
    # Check if running cleanup mode
    if len(sys.argv) > 1 and sys.argv[1] == '--cleanup':
        schedule_midnight_cleanup()
        driver.close()
        sys.exit(0)
    
    # Setup (run once)
    if len(sys.argv) > 1 and sys.argv[1] == '--setup':
        create_cache_node_constraint(driver)
        driver.close()
        sys.exit(0)
    
    # Normal execution
    tickers = ['WMT', 'SPB', 'SFM']
    
    results = calculate_metrics_for_tickers(tickers)
    formatted_results = format_output(results)
    
    # Print results
    print("\n" + "="*80)
    print("STOCK RANKING RESULTS")
    print("="*80)
    for item in formatted_results:
        print(f"\nOverall Rank: {item['Overall Rank']}")
        print(f"Ticker: {item['Ticker']}")
        print(f"ROIC 5Y Avg: {item['ROIC 5Y Avg']}")
        print(f"Earnings Yield: {item['Earnings Yield']}")
        print(f"Intrinsic to Market Cap: {item['Intrinsic to Market Cap']}")
        print("-"*80)
    
    driver.close()
```
