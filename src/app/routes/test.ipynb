```python
import pandas as pd

def csv_to_markdown(csv_path, output_path):
    """
    Convert a CSV file to a Markdown table and save it to a text file.

    :param csv_path: Path to the input CSV file
    :param output_path: Path to the output TXT file
    """
    # Read the CSV into a pandas DataFrame
    df = pd.read_csv(csv_path)
    
    # Convert DataFrame to Markdown format
    markdown_table = df.to_markdown(index=False)
    
    # Write the Markdown to a text file
    with open(output_path, "w", encoding="utf-8") as f:
        f.write(markdown_table)
    
    print(f"Markdown table saved to: {output_path}")

# Example usage:
csv_path = "/Users/dimuth/Documents/chat_bot/data_given_anand/stage_5/MultiplesData/ADBE_MultiplesTable.csv"        # Replace with your CSV path
output_path = "output_markdown.txt"  # Replace with desired output path
csv_to_markdown(csv_path, output_path)
```

```python
import requests
from bs4 import BeautifulSoup
from markdownify import markdownify as md
import requests

HEADERS = {
    "User-Agent": "YourName your@email.com"
}

TICKER = "AGCO"
YEAR = "2021"

# --------------------------------------------------
# 1) Ticker → CIK
# --------------------------------------------------
ticker_url = "https://www.sec.gov/files/company_tickers.json"
ticker_data = requests.get(ticker_url, headers=HEADERS).json()

cik = None
for item in ticker_data.values():
    if item["ticker"].upper() == TICKER.upper():
        cik = str(item["cik_str"]).zfill(10)
        break

if not cik:
    raise ValueError("Ticker not found")

print(f"CIK for {TICKER}: {cik}")

# --------------------------------------------------
# 2) Get company filings
# --------------------------------------------------
submissions_url = f"https://data.sec.gov/submissions/CIK{cik}.json"
data = requests.get(submissions_url, headers=HEADERS).json()

forms = data["filings"]["recent"]["form"]
dates = data["filings"]["recent"]["filingDate"]
docs = data["filings"]["recent"]["primaryDocument"]
accessions = data["filings"]["recent"]["accessionNumber"]

# --------------------------------------------------
# 3) Find 10-K for the requested year
# --------------------------------------------------
found = False

for i in range(len(forms)):
    if forms[i] == "10-K" and dates[i].startswith(YEAR):
        acc = accessions[i].replace("-", "")
        doc = docs[i]
        html_url = f"https://www.sec.gov/Archives/edgar/data/{int(cik)}/{acc}/{doc}"
        print(f"10-K {YEAR} URL:")
        print(html_url)
        found = True
        break

if not found:
    print(f"No 10-K found for {TICKER} in {YEAR}")




#-----------------------#
#-----------------------#


headers = {"User-Agent": "YourName your@email.com"}
html = requests.get(html_url, headers=headers).text

with open(f"{TICKER}_{YEAR}_10K.html", "w", encoding="utf-8") as f:
    f.write(html)

soup = BeautifulSoup(html, "html.parser")
for tag in soup(["script", "style"]):
    tag.decompose()
markdown = md(str(soup), heading_style="ATX")

with open(f"{TICKER}_{YEAR}_10K.md", "w", encoding="utf-8") as f:
    f.write(markdown)
```

```python
import pandas as pd
import numpy as np


def run_v_elite_engine_v2(df: pd.DataFrame) -> pd.DataFrame:
    """
    The Definitive VInvest Engine (DataFrame Version)

    - Accepts a pre-loaded DataFrame (CSV, DB, API, etc.)
    - Calculates all sub-ratings
    - Applies the Gatekeeper
    - Assigns nested Elite Tiers
    """

    # --------------------------------------------------
    # STEP 0: DEFENSIVE COPY + VALIDATION
    # --------------------------------------------------
    df = df.copy()

    if 'Ticker' not in df.columns:
        raise ValueError("Input DataFrame must contain a 'Ticker' column")

    required_cols = [
        'ReturnOnInvestedCapital',
        'RevenueGrowth',
        'ShareDilution',
        
        'EarningsYield',
        'IntrinsicToMarketCap',
        'AltmanZScore',
        'PiotroskiFScore',
        'DebtToEBITDA',
        'ROC_6M',
        'Above_200SMA',
        'RSI_14'
    ]

    missing = set(required_cols) - set(df.columns)
    if missing:
        raise ValueError(f"Missing required columns: {missing}")

    # --------------------------------------------------
    # STEP 1: CALCULATE UNIVERSAL SUB-RATINGS (0–100)
    # --------------------------------------------------

    # Pillar 1: V-Quality
    df['V_Quality'] = (
        df['ReturnOnInvestedCapital'].rank(pct=True) * 0.4 +
        df['RevenueGrowth'].rank(pct=True) * 0.4 +
        (1 - df['ShareDilution'].rank(pct=True)) * 0.2
    ) * 100

    # Pillar 2: V-Value
    df['V_Value'] = (
        df['EarningsYield'].rank(pct=True) * 0.5 +
        df['IntrinsicToMarketCap'].rank(pct=True) * 0.5
    ) * 100

    # Pillar 3: V-Safety
    df['V_Safety'] = (
        df['AltmanZScore'].rank(pct=True) * 0.4 +
        (df['PiotroskiFScore'] / 9) * 0.4 +
        (1 - df['DebtToEBITDA'].rank(pct=True)) * 0.2
    ) * 100

    # Pillar 4: V-Momentum
    rsi_sig = np.where(
        (df['RSI_14'] > 30) & (df['RSI_14'] < 65),
        1.0,
        0.5
    )

    df['V_Momentum'] = (
        df['ROC_6M'].rank(pct=True) * 0.4 +
        df['Above_200SMA'] * 0.4 +
        rsi_sig * 0.2
    ) * 100

    # --------------------------------------------------
    # STEP 2: CALCULATE AGGREGATE V-RATING
    # --------------------------------------------------
    df['V_Rating'] = (
        df['V_Quality'] * 0.30 +
        df['V_Value'] * 0.30 +
        df['V_Safety'] * 0.25 +
        df['V_Momentum'] * 0.15
    ).round(2)

    # --------------------------------------------------
    # STEP 3: GATEKEEPER WATERFALL
    # --------------------------------------------------
    is_distressed = df['AltmanZScore'] <= 1.1
    is_overleveraged = df['DebtToEBITDA'] >= 6.0
    is_unprofitable = df['ReturnOnInvestedCapital'] <= 0

    conditions = [
        is_distressed,
        is_overleveraged,
        is_unprofitable
    ]

    rejection_labels = [
        "REJECT: Distress Zone",
        "REJECT: Over-Leveraged",
        "REJECT: Zero/Neg ROIC"
    ]

    df['Status'] = np.select(
        conditions,
        rejection_labels,
        default="Qualified"
    )

    # --------------------------------------------------
    # STEP 4: NESTED ELITE TIERING (Qualified Only)
    # --------------------------------------------------
    df['Rank'] = np.nan

    qualified_mask = df['Status'] == "Qualified"

    ranked = (
        df.loc[qualified_mask]
        .sort_values(
            by=['V_Rating', 'IntrinsicToMarketCap'],
            ascending=False
        )
        .reset_index()
    )

    ranked['Rank'] = ranked.index + 1

    df.loc[qualified_mask, 'Rank'] = ranked['Rank'].values

    def assign_tier(rank):
        if pd.isna(rank):
            return None
        if rank <= 10:
            return "V-Elite 10"
        if rank <= 50:
            return "V-Elite 50"
        if rank <= 100:
            return "V-Elite 100"
        if rank <= 500:
            return "V-Elite 500"
        return "V-Qualified"

    df['Status'] = np.where(
        df['Status'] == "Qualified",
        df['Rank'].apply(assign_tier),
        df['Status']
    )

    # --------------------------------------------------
    # STEP 5: FINAL OUTPUT
    # --------------------------------------------------
    output_cols = [
        'Ticker',
        'V_Rating',
        'Status',
        'Rank',
        'V_Quality',
        'V_Value',
        'V_Safety',
        'V_Momentum'
    ]

    return (
        df[output_cols]
        .sort_values(
            by=['Rank', 'V_Rating'],
            ascending=[True, False],
            na_position='last'
        )
        .reset_index(drop=True)
    )


# --------------------------------------------------
# EXAMPLE USAGE
# --------------------------------------------------
if __name__ == "__main__":
    v_registry = pd.read_csv("/Users/dimuth/Documents/chat_bot/Get-Deep/src/app/routes/vinvest_process_log_shared.csv")
    final_tracker_df = run_v_elite_engine_v2(v_registry)
    print(final_tracker_df.head(20))
```

```python
# --- PILLAR 2: V-VALUE ---
# of that VEliteYield - 50% and IntrinsicToMarketCap - 50%
#V-Elite Yield	EBIT / Market Enterprise Value	The Truth. It shows how much the actual business generates regardless of how it's funded.
#MarketEnterpriseValue = MarketCap + Debt + LeaseLiabilities - ExcessCash
v_registry[ticker]['MarketEnterpriseValue'] = MarketCap + ValuationSummaryTable.at['Debt','PresentValue'] + ValuationSummaryTable.at['OperatingLeaseLiabilities','PresentValue'] + ValuationSummaryTable.at['FinanceLeaseLiabilities','PresentValue'] + ValuationSummaryTable.at['VariableLeaseLiabilities','PresentValue'] - ValuationSummaryTable.at['ExcessCash','PresentValue']
v_registry[ticker]['VEliteYield'] = IncomeStatement.at['EBIT', latest_filing_year_str] / v_registry[ticker]['MarketEnterpriseValue']

v_registry[ticker]['IntrinsicToMarketCap'] = ValuationSummaryTable.at['EquityIntrinsicValue','PresentValue'] / MarketCap






# --- PILLAR 3: V-SAFETY ---
#Altman Z-Score - 40%, Piotroski F-Score - 40%, Debt/EBITDA - 20%

#Altman Z-Score
#X1: Working Capital / Total Assets (Liquidity)
#X2: Retained Earnings / Total Assets (Accumulated Profitability)
#X3: EBIT / Total Assets (Operating Efficiency)
#X4: Market Value of Equity / Total Liabilities (Solvency/Leverage)
#X5: Sales / Total Assets (Asset Turnover)
#NOWC = (CurrentAssets - Cash) - (CurrentLiabilities - ShortTermDebt)
v_registry[ticker]['ZScoreLiquidity'] = HistoricalCagrAvg.at['NetOperatingWorkingCapital', 'Last1Y_AVG'] / HistoricalCagrAvg.at['Assets', 'Last1Y_AVG']
v_registry[ticker]['ZScoreAccumulatedProfitability'] = HistoricalCagrAvg.at['RetainedEarningsAccumulated', 'Last1Y_AVG'] / HistoricalCagrAvg.at['Assets', 'Last1Y_AVG']
v_registry[ticker]['ZScoreOperatingEfficiency'] = HistoricalCagrAvg.at['EBIT', 'Last1Y_AVG'] / HistoricalCagrAvg.at['Assets', 'Last1Y_AVG']
v_registry[ticker]['ZScoreSolvency'] = MarketCap / HistoricalCagrAvg.at['Liabilities', 'Last1Y_AVG']
v_registry[ticker]['ZScoreAssetTurnover'] = HistoricalCagrAvg.at['Revenue', 'Last1Y_AVG'] / HistoricalCagrAvg.at['Assets', 'Last1Y_AVG']

x1 = v_registry[ticker]['ZScoreLiquidity']
x2 = v_registry[ticker]['ZScoreAccumulatedProfitability']
x3 = v_registry[ticker]['ZScoreOperatingEfficiency']
x4 = v_registry[ticker]['ZScoreSolvency']
x5 = v_registry[ticker]['ZScoreAssetTurnover']

z_score = (1.2 * x1) + (1.4 * x2) + (3.3 * x3) + (0.6 * x4) + (0.999 * x5)
v_registry[ticker]['AltmanZScore'] = round(z_score, 2)
#Piotroski F-Score

#Calculates the 0-9 F-Score for a single company.
#'curr' and 'prev' are dictionaries or Series containing the financial keys.
prior_year_to_latest_filing_year_str = str(latest_filing_year - 1)
f_score = 0
try:
    # --- CATEGORY 1: PROFITABILITY (4 pts) ---
    # 1. ROA > 0
    roa_curr = IncomeStatement.at['NetIncome',latest_filing_year_str] / BalanceSheet.at['Assets',latest_filing_year_str]
    if roa_curr > 0: f_score += 1
    
    # 2. Positive Operating Cash Flow
    if CashFlow.at['OperatingCashFlow',latest_filing_year_str] > 0: f_score += 1
    
    # 3. ROA Trend (Is current ROA > previous ROA?)
    roa_prev = IncomeStatement.at['NetIncome',prior_year_to_latest_filing_year_str] / BalanceSheet.at['Assets',prior_year_to_latest_filing_year_str]
    if roa_curr > roa_prev: f_score += 1
    
    # 4. Accruals (Quality of Earnings)
    # Point if Cash Flow > Net Income
    if CashFlow.at['OperatingCashFlow',latest_filing_year_str] > IncomeStatement.at['NetIncome',latest_filing_year_str]: f_score += 1

    # --- CATEGORY 2: LEVERAGE & LIQUIDITY (3 pts) ---
    # 5. Lower Leverage (Long-term Debt / Total Assets)
    lev_curr = BalanceSheet.at['LongTermDebt',latest_filing_year_str] / BalanceSheet.at['Assets',latest_filing_year_str]
    lev_prev = BalanceSheet.at['LongTermDebt',prior_year_to_latest_filing_year_str] / BalanceSheet.at['Assets',prior_year_to_latest_filing_year_str]
    if lev_curr < lev_prev: f_score += 1
    
    # 6. Higher Liquidity (Current Ratio)
    cr_curr = BalanceSheet.at['AssetsCurrent',latest_filing_year_str] / BalanceSheet.at['LiabilitiesCurrent',latest_filing_year_str]
    cr_prev = BalanceSheet.at['AssetsCurrent',prior_year_to_latest_filing_year_str] / BalanceSheet.at['LiabilitiesCurrent',prior_year_to_latest_filing_year_str]
    if cr_curr > cr_prev: f_score += 1
    
    # 7. No Dilution (Shares Outstanding)
    # Note: We give a point if shares stayed the same or decreased
    SharesOutstanding = 1000000 #API call
    PrioYearSharesOutstanding = 1000000 #API call
    if SharesOutstanding <= PrioYearSharesOutstanding: f_score += 1

    # --- CATEGORY 3: OPERATING EFFICIENCY (2 pts) ---
    # 8. Gross Margin Trend
    gm_curr = IncomeStatement.at['GrossMargin',latest_filing_year_str] / IncomeStatement.at['Revenue',latest_filing_year_str]
    gm_prev = IncomeStatement.at['GrossMargin',prior_year_to_latest_filing_year_str] / IncomeStatement.at['Revenue',prior_year_to_latest_filing_year_str]
    if gm_curr > gm_prev: f_score += 1
    
    # 9. Asset Turnover Trend
    at_curr = IncomeStatement.at['Revenue',latest_filing_year_str] / BalanceSheet.at['Assets',latest_filing_year_str]
    at_prev = IncomeStatement.at['Revenue',prior_year_to_latest_filing_year_str] / BalanceSheet.at['Assets',prior_year_to_latest_filing_year_str]
    if at_curr > at_prev: f_score += 1
except: 
    print(f"Missing F-Score calculation.")

v_registry[ticker]['PiotroskiFScore'] = f_score













# --- PILLAR 4: V-MOMENTUM (Including RSI) ---

#Calculates the V-Momentum components for a single ticker.
#'price_history' should be a list or array of the last 200 closing prices.
#Order: [Oldest ... Latest]

#if not price history return {"ROC": 0, "SMA_Signal": 0, "RSI": 50, "Status": "Incomplete Data"}
price_history_file = f'{data_folder_path}/COST-PriceHistory.csv'
stock_history = pd.read_csv(price_history_file)
price_history = stock_history['Close'].tolist()
#For your V-Momentum 200-day check, get last 200 items
if len(price_history) >= 200:
    price_history = price_history[-200:]
else:
    price_history = price_history  # Use whatever is available
latest_price = price_history[-1]

# 1. Rate of Change (6-Month ROC)
# We use ~126 trading days for 6 months
price_6m_ago = price_history[-126]
roc_6m = ((latest_price - price_6m_ago) / price_6m_ago) * 100

# 2. Trend Confirmation (Price vs 200-Day SMA)
sma_200 = np.mean(price_history)
is_above_sma = 1 if latest_price > sma_200 else 0

# 3. Relative Strength Index (14-Day RSI)
# Measures if the stock is 'Overbought' or 'Oversold'
deltas = np.diff(price_history[-15:]) # Need 15 days for 14 deltas
gains = deltas[deltas > 0]
losses = -deltas[deltas < 0]

avg_gain = np.mean(gains) if len(gains) > 0 else 0
avg_loss = np.mean(losses) if len(losses) > 0 else 1e-9 # Prevent div by zero

rs = avg_gain / avg_loss
rsi_14 = 100 - (100 / (1 + rs))

v_registry[ticker]['ROC_6M'] = round(roc_6m, 2)
v_registry[ticker]['Above_200SMA'] = is_above_sma
v_registry[ticker]['RSI_14'] = round(rsi_14, 2)











import pandas as pd
import numpy as np

def run_v_elite_engine_v2(v_registry_dict):
    """
    The Definitive VInvest Engine: Calculates all sub-ratings,
    applies the Gatekeeper, and assigns nested Elite Tiers.
    """
    # 1. Convert dictionary to DataFrame
    df = pd.DataFrame(v_registry_dict).T
    df.index.name = 'Ticker'
    df = df.reset_index()

    # --- STEP 1: CALCULATE UNIVERSAL SUB-RATINGS (0-100) ---
    # We calculate these for EVERY ticker to allow for deep-dive analysis.
    
    # Pillar 1: V-Quality
    df['V_Quality'] = (
        df['ReturnOnInvestedCapital'].rank(pct=True) * 0.4 +
        df['RevenueGrowth'].rank(pct=True) * 0.4 +
        (1 - df['ShareDilution'].rank(pct=True)) * 0.2
    ) * 100

    # Pillar 2: V-Value
    df['V_Value'] = (
        df['VEliteYield'].rank(pct=True) * 0.5 +
        df['IntrinsicToMarketCap'].rank(pct=True) * 0.5
    ) * 100

    # Pillar 3: V-Safety
    df['V_Safety'] = (
        df['AltmanZScore'].rank(pct=True) * 0.4 +
        (df['PiotroskiFScore'] / 9) * 0.4 +
        (1 - df['DebtToEBITDA'].rank(pct=True)) * 0.2
    ) * 100

    # Pillar 4: V-Momentum
    rsi_sig = df['RSI_14'].apply(lambda x: 1.0 if 30 < x < 65 else 0.5)
    df['V_Momentum'] = (
        df['ROC_6M'].rank(pct=True) * 0.4 +
        df['Above_200SMA'] * 0.4 +
        rsi_sig * 0.2
    ) * 100

    # --- STEP 2: CALCULATE AGGREGATE V-RATING ---
    df['V_Rating'] = (
        df['V_Quality'] * 0.30 +
        df['V_Value'] * 0.30 +
        df['V_Safety'] * 0.25 +
        df['V_Momentum'] * 0.15
    ).round(2)

    # --- STEP 3: THE GATEKEEPER WATERFALL ---
    is_distressed = df['AltmanZScore'] <= 1.1
    is_overleveraged = df['DebtToEBITDA'] >= 6.0
    is_unprofitable = df['ReturnOnInvestedCapital'] <= 0
    
    conditions = [is_distressed, is_overleveraged, is_unprofitable]
    rejection_labels = [
        "REJECT: Distress Zone", 
        "REJECT: Over-Leveraged", 
        "REJECT: Zero/Neg ROIC"
    ]
    
    # Initial status assignment
    df['Status'] = np.select(conditions, rejection_labels, default="Qualified")
    
    # --- STEP 4: NESTED TIERING FOR QUALIFIED TICKERS ---
    # We only rank the ones that passed the Gatekeeper
    qualified_mask = df['Status'] == "Qualified"
    
    # Sort survivors by Rating and then Logic Gap (tie-breaker)
    df.loc[qualified_mask, 'Rank'] = (
        df[qualified_mask]
        .sort_values(by=['V_Rating', 'IntrinsicToMarketCap'], ascending=False)
        .reset_index()
        .index + 1
    )

    # Assign Elite Tiers based on Rank
    def assign_tier(r):
        if pd.isna(r): return None
        if r <= 10: return "V-Elite 10"
        if r <= 50: return "V-Elite 50"
        if r <= 100: return "V-Elite 100"
        if r <= 500: return "V-Elite 500"
        return "V-Qualified"

    # Update Status for those that passed
    tier_labels = df['Rank'].apply(assign_tier)
    df['Status'] = np.where(df['Status'] == "Qualified", tier_labels, df['Status'])

    # --- STEP 5: FINAL CLEANUP & OUTPUT ---
    cols = [
        'Ticker', 'V_Rating', 'Status', 'Rank',
        'V_Quality', 'V_Value', 'V_Safety', 'V_Momentum'
    ]
    
    # Sort so the Elite 10 are at the top, followed by others, then Rejects at bottom
    return df[cols].sort_values(by=['Rank', 'V_Rating'], ascending=[True, False], na_position='last')

# Example Usage:
v_registry = pd.read_csv()
final_tracker_df = run_v_elite_engine_v2(v_registry)
```

```python
import yfinance as yf
import pandas as pd

def get_shares_outstanding_ago(ticker_symbol, years_ago=1):
    ticker = yf.Ticker(ticker_symbol)
    shares_hist = ticker.get_shares_full()
    if shares_hist.empty:
        raise ValueError(f"No shares outstanding data found for {ticker_symbol}")
    shares_hist.index = pd.to_datetime(shares_hist.index).tz_localize(None)
    target_date = pd.Timestamp.today() - pd.DateOffset(years=years_ago)
    filtered = shares_hist.loc[shares_hist.index <= target_date]
    if filtered.empty:
        raise ValueError(f"No shares outstanding data available before {target_date.date()}")
    shares_ago = filtered.iloc[-1]
    if isinstance(shares_ago, pd.Series):
        return int(shares_ago.values[0])
    else:
        return int(shares_ago.item())

# shares_1y_ago = get_shares_outstanding_ago("COST", years_ago=1)
# print("Shares outstanding ~1 year ago:", shares_1y_ago)
```

```python
import yfinance as yf

def get_last_60d_closing_prices(ticker_symbol):
    """
    Returns last 60 days of closing prices for a stock as a list,
    ordered from latest to oldest.
    """
    data = yf.download(ticker_symbol,period="60d",interval="1d",progress=False)
    closing_prices = data[["Close"]].squeeze()  # <- converts DataFrame to Series if needed
    closing_prices = closing_prices[::-1]
    return closing_prices.tolist()

prices_list = get_last_60d_closing_prices("BIGC")
print(prices_list)
```
